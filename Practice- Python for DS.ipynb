# Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
fig = plt.figure()
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(y,x)
#axes[2].plot(x,y)
#axes[3].plot(x,y)
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure()
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0,0].plot(x,y)
axes[0,1].plot(y,x)
axes[1,0].plot(x,y)
axes[1,1].plot(x,y)
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
fig=plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
```

```python
ax1.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')

fig # to show figure object
```

```python
# plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]

fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

```python
#xlimits and y limits on the inserted plot

ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2.plot(x,y)
ax2.set_title('zoom')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

fig
```

```python
fig = plt.figure()
#crete empty canvas
fig,axes = plt.subplots(nrows=1, ncols=2)
```

```python
axes[0].plot(x,y, ls='--',color='blue',lw=3)
axes[1].plot(x,z,color='red',lw=3)


fig
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))

axes[0].plot(x,y,color="blue", lw=5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z,color="red", lw=3, ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=40)
```

```python
sns.distplot(tips['sex']) #only number datatype is used for distribution plot
```

```python
sns.jointplot(x='total_bill',y='tip', data=tips, kind='hex') # join 2 distribution plot
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```
